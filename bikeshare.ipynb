## Create dataframes for all four files

```python
path = "/FileStore/tables/riders.csv"
df_riders = spark.read \
    .option("inferSchema", "false") \
    .option("header", "false") \
    .format("csv") \
    .option("sep", ",") \
    .load(path) 

display(df_riders)
```

```python
path = "/FileStore/tables/payments.csv"
df_payments = spark.read \
    .option("inferSchema", "false") \
    .option("header", "false") \
    .format("csv") \
    .option("sep", ",") \
    .load(path) 

display(df_payments)
```

```python
path = "/FileStore/tables/stations.csv"
df_stations = spark.read \
    .option("inferSchema", "false") \
    .option("header", "false") \
    .format("csv") \
    .option("sep", ",") \
    .load(path) 

display(df_stations)
```

```python
path = "/FileStore/tables/trips.csv"
df_trips = spark.read \
    .option("inferSchema", "false") \
    .option("header", "false") \
    .format("csv") \
    .option("sep", ",") \
    .load(path) 

display(df_trips)
```

## Write all your data out to delta

```python
df_riders.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("riders")

df_stations.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("stations")

df_payments.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("payments")

df_trips.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("trips")
```
